# Topic modeling

```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation as LDA
import pickle as pic
from sklearn.feature_extraction.text import TfidfVectorizer
import os
from pyLDAvis import sklearn as sklearn_lda
import pickle 
import pyLDAvis
```

## Reading the data

```python
df = pd.read_csv('cleandata.csv')  
df = df.dropna()
df.head()
```

Allowing visualizations in the notebook. An initial number of topics has been chosen in order to create a first visualization

```python
pyLDAvis.enable_notebook()
number_topics= 5
```

## Topic modeling for English

```python
dfaux = df[df["label_name"]=='en']
vectorizer = TfidfVectorizer()
vectorizer.fit(dfaux["text"].values)
count_data = vectorizer.transform(dfaux["text"].values)
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
LDAvis_prepared = sklearn_lda.prepare(lda, count_data, vectorizer)
pyLDAvis.display(LDAvis_prepared )
```

## Topic modeling for Spanish

```python
dfaux = df[df["label_name"]=='es']
vectorizer = TfidfVectorizer()
vectorizer.fit(dfaux["text"].values)
count_data = vectorizer.transform(dfaux["text"].values)
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
LDAvis_prepared = sklearn_lda.prepare(lda, count_data, vectorizer)
pyLDAvis.display(LDAvis_prepared )
```

## Topic modeling for French

```python
dfaux = df[df["label_name"]=='fr']
vectorizer = TfidfVectorizer()
vectorizer.fit(dfaux["text"].values)
count_data = vectorizer.transform(dfaux["text"].values)
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
LDAvis_prepared = sklearn_lda.prepare(lda, count_data, vectorizer)
pyLDAvis.display(LDAvis_prepared )
```

## Conclusions


As it can be appreciated from the representations in all the three languages that we had on the dataset, there are several documents that contained, if are not whole made up of, terms that are not from the language they were supposed to be. This makes the topic extraction come up with topics associated with languages, as the main topic is the selected language and the rest of the small topics are words that do not belong in the same structure as the main topic.

With some time, a more thorough cleaning and exploration process would have determined this in advance and would have cleaned this inconsistencies. Afterwards with the documents cleaned of those special cases, it could have been spected that a more explicative set of topics could have been achieved.
